Scatter plots are the graphs that present the relationship between two variables in a data-set. It represents data points on a two-dimensional plane or on a Cartesian system. The independent variable or attribute is plotted on the X-axis, while the dependent variable is plotted on the Y-axis. These plots are often called scatter graphs or scatter diagrams.


```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y, color='red')
plt.title('Scatter plot of random generated data')
plt.xlabel('X-axis labels')
plt.ylabel('Y-axis labels')
```

#Q3

The subplot() function in Matplotlib is used to create multiple plots (also known as subplots) within a single figure. This is particularly useful when you want to compare different data visualizations side by side or when you have multiple related plots that you want to present together in a grid layout

```python
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.plot([0, 1, 2, 3, 4, 5],[0, 100, 200, 300, 400, 500])
plt.title("1st plot")

plt.subplot(2,2,2)
plt.plot([0, 1, 2, 3, 4, 5],[50, 20, 40, 20, 60, 70])
plt.title("2nd plot")

plt.subplot(2,2,3)
plt.plot([0, 1, 2, 3, 4, 5],[10, 20, 30, 40, 50, 60])
plt.title("3rd plot")

plt.subplot(2,2,4)
plt.plot([0, 1, 2, 3, 4, 5],[200, 350, 250, 550, 450, 150])
plt.title("4th plot")
```

#Q4

Bar plots are used primarily for the following reasons:

Comparison of Categories: They provide a clear visual comparison between different categories or groups, making it easy to see which categories have higher or lower values.

Simple Visualization: Bar plots are straightforward and easy to interpret, making them accessible for a wide audience.

Trend Identification: They help in identifying trends over time when comparing data across different periods.

Frequency Distribution: Bar plots can effectively show the frequency or count of categories within a dataset.

Highlighting Differences: They are useful for emphasizing significant differences or outliers among categories.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
plt.title('Company Profits (Vertical Bar Plot)')
plt.xlabel('Company names')
plt.ylabel('Company profit')
```

```python
plt.barh(company, profit, color='orange')
plt.title('Company Profits (Horizontal Bar Plot)')
plt.xlabel('Profit in USD')
plt.ylabel('Company')
```
